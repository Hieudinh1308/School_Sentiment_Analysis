```python
def predict_sentiment (dataframe , topic) :
  predicted = []
  for x in dataframe[topic]:
    sentiment_lecturer_topic	= model(str(x))
    predicted.append(sentiment_lecturer_topic)
  return predicted
```

```python
df_pedicted = pd.DataFrame()
df_pedicted['lecturer_topic_predict'] = predict_sentiment(df,"lecturer_topic")
df_pedicted['learning_program_topic_predict'] = predict_sentiment(df,"learning_program_topic")
df_pedicted['facility_topic_predict'] = predict_sentiment(df,"facility_topic")
df_pedicted['orther_topic_predict'] = predict_sentiment(df,"orther_topic")
df_pedicted[:]
```

```python
# analysis

def analysis (df_pedicted, topic) :

  label_predicted = []
  label_POSITIVE = label_NEGATIVE  = 0
  for i in df_pedicted[topic] :
    label_predicted.append(i[0]['label'])
    if i[0]['label'] == "POSITIVE" :
      label_POSITIVE +=1
    elif i[0]['label'] == "NEGATIVE" :
      label_NEGATIVE +=1

  print( f" \
  total examples for {topic} topic {len(label_predicted)} \n \
  the 5 first predicted lables {label_predicted[:5]} \
  lenght of pos label {label_POSITIVE} \
  lenght of neg label {label_NEGATIVE} \
  lenghe of neu label {len(label_predicted) -  label_POSITIVE - label_NEGATIVE} \
        \n ")

  return len(label_predicted), label_predicted, \
        label_POSITIVE,label_NEGATIVE, \
        len(label_predicted) -  label_POSITIVE - label_NEGATIVE
```

```python
analysis(df_pedicted ,'lecturer_topic_predict')
```

```python
df_pedicted[:1]
```

```python
_,_,lecturer_topic_predict_pos,lecturer_topic_predict_neg,lecturer_topic_predict_neu=analysis(df_pedicted ,'lecturer_topic_predict')

_,_,learning_program_topic_predict_pos,learning_program_topic_predict_neg,learning_program_topic_predict_neu=analysis(df_pedicted ,'learning_program_topic_predict')

_,_,facility_topic_predict_pos,facility_topic_predict_neg,facility_topic_predict_neu=analysis(df_pedicted ,'facility_topic_predict')

_,_,orther_topic_predict_pos,orther_topic_predict_neg,orther_topic_predict_neu=analysis(df_pedicted ,'orther_topic_predict')
```

```python
def daw_pie (pos,neg,neu,tile):
  import matplotlib.pyplot as plt

  # Lengths of the labels
  length_positive = pos
  length_negative = neg
  length_neutral = neu

  # Data for the pie chart
  lengths = [length_positive, length_negative, length_neutral]
  labels = ['Positive', 'Negative', 'Neutral']
  colors = ['green', 'red', 'blue']

  plt.figure(figsize=(8, 6))
  plt.pie(lengths, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
  plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
  plt.title(f'Proportions of Sentiment Label {tile} Lengths')

  plt.show()
```

```python
daw_pie(lecturer_topic_predict_pos,lecturer_topic_predict_neg,lecturer_topic_predict_neu,"lecturer_topic")
print("\n")
daw_pie(learning_program_topic_predict_pos,learning_program_topic_predict_neg,learning_program_topic_predict_neu,"learning_program_topic")
print("\n")
daw_pie(facility_topic_predict_pos,facility_topic_predict_neg,facility_topic_predict_neu,"facility_topic")
print("\n")
daw_pie(orther_topic_predict_pos,orther_topic_predict_neg,orther_topic_predict_neu,"orther_topic")
```
